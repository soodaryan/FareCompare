# Transit Routing System
Build a system to find nearest bus stops from coordinates and find routes between them

```python
from math import radians, cos, sin, asin, sqrt

# Function to calculate haversine distance between two points
def haversine_distance(lat1, lon1, lat2, lon2):
    """
    Calculate the great circle distance between two points 
    on the earth (specified in decimal degrees)
    Returns distance in kilometers
    """
    # Convert decimal degrees to radians
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    
    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    r = 6371  # Radius of earth in kilometers
    
    return c * r

# Function to find the nearest stop to given coordinates
def find_nearest_stop(lat, lon, stops_df, top_n=5):
    """
    Find the nearest bus stops to given coordinates
    
    Parameters:
    - lat, lon: latitude and longitude of the location
    - stops_df: dataframe containing stops information (df2)
    - top_n: number of nearest stops to return
    
    Returns: DataFrame with nearest stops
    """
    # Calculate distance from the input coordinates to all stops
    stops_df['distance_km'] = stops_df.apply(
        lambda row: haversine_distance(lat, lon, row['stop_lat'], row['stop_lon']),
        axis=1
    )
    
    # Sort by distance and return top N stops
    nearest_stops = stops_df.nsmallest(top_n, 'distance_km')[['stop_id', 'stop_name', 'stop_lat', 'stop_lon', 'distance_km']]
    return nearest_stops

# Test the function with sample coordinates (Delhi coordinates)
sample_lat, sample_lon = 28.7041, 77.1025  # New Delhi center
print("Finding nearest stops to Delhi center coordinates...")
nearest = find_nearest_stop(sample_lat, sample_lon, df2.copy())
print(nearest)
```

```python
# Function to find routes between two stops
def find_routes_between_stops(start_stop_id, end_stop_id, stop_times_df, trips_df, routes_df):
    """
    Find all possible routes that connect two stops
    
    Parameters:
    - start_stop_id, end_stop_id: stop IDs for start and end locations
    - stop_times_df: dataframe with stop times (df1)
    - trips_df: dataframe with trips (df8)
    - routes_df: dataframe with routes (df4)
    
    Returns: List of route_ids and corresponding route details
    """
    # Find all trips that pass through the starting stop
    trips_at_start = stop_times_df[stop_times_df['stop_id'] == start_stop_id]['trip_id'].unique()
    
    # Find all trips that pass through the ending stop
    trips_at_end = stop_times_df[stop_times_df['stop_id'] == end_stop_id]['trip_id'].unique()
    
    # Find trips that pass through both stops
    common_trips = set(trips_at_start) & set(trips_at_end)
    
    if len(common_trips) == 0:
        print(f"No direct routes found from stop {start_stop_id} to stop {end_stop_id}")
        return pd.DataFrame()
    
    # IMPORTANT: Filter trips where start_stop comes BEFORE end_stop in the sequence
    valid_trips = []
    for trip_id in common_trips:
        trip_stops = stop_times_df[stop_times_df['trip_id'] == trip_id]
        start_seq = trip_stops[trip_stops['stop_id'] == start_stop_id]['stop_sequence'].values
        end_seq = trip_stops[trip_stops['stop_id'] == end_stop_id]['stop_sequence'].values
        
        if len(start_seq) > 0 and len(end_seq) > 0:
            if start_seq[0] < end_seq[0]:  # Start comes before end in the sequence
                valid_trips.append(trip_id)
    
    if len(valid_trips) == 0:
        print(f"No routes found where stop {start_stop_id} comes before stop {end_stop_id}")
        return pd.DataFrame()
    
    # Get route information for these valid trips
    routes_for_trips = trips_df[trips_df['trip_id'].isin(valid_trips)][['trip_id', 'route_id']].drop_duplicates()
    
    # Merge with route details
    result = routes_for_trips.merge(routes_df, on='route_id', how='left')
    
    return result

# Test the function with two sample stops
print("\n" + "="*50)
print("Finding routes between stops...")
sample_start_stop = df1['stop_id'].iloc[0]
sample_end_stop = df1['stop_id'].iloc[100]  # Different stop

print(f"Start stop: {sample_start_stop}, End stop: {sample_end_stop}")
routes = find_routes_between_stops(sample_start_stop, sample_end_stop, df1, df8, df4)
print("\nAvailable routes:")
print(routes[['trip_id', 'route_id', 'route_short_name', 'route_long_name']] if len(routes) > 0 else "No routes found")
```

```python
# Comprehensive function to get the complete journey
def plan_journey(start_lat, start_lon, end_lat, end_lon, stops_df, stop_times_df, trips_df, routes_df):
    """
    Complete journey planner:
    1. Find nearest stop to start coordinates
    2. Find nearest stop to end coordinates
    3. Find all routes connecting these stops
    
    Parameters:
    - start_lat, start_lon: Starting coordinates
    - end_lat, end_lon: Ending coordinates
    - stops_df, stop_times_df, trips_df, routes_df: GTFS dataframes
    
    Returns: Dictionary with journey information
    """
    print("="*60)
    print("JOURNEY PLANNER")
    print("="*60)
    
    # Find nearest stops
    print(f"\n1. Finding nearest bus stop to start location ({start_lat}, {start_lon})...")
    start_stops = find_nearest_stop(start_lat, start_lon, stops_df.copy(), top_n=3)
    print(start_stops.to_string())
    
    print(f"\n2. Finding nearest bus stop to end location ({end_lat}, {end_lon})...")
    end_stops = find_nearest_stop(end_lat, end_lon, stops_df.copy(), top_n=3)
    print(end_stops.to_string())
    
    # Get the nearest stop from each location
    nearest_start_stop = start_stops.iloc[0]['stop_id']
    nearest_end_stop = end_stops.iloc[0]['stop_id']
    
    print(f"\n3. Finding routes from stop {nearest_start_stop} to stop {nearest_end_stop}...")
    routes = find_routes_between_stops(nearest_start_stop, nearest_end_stop, stop_times_df, trips_df, routes_df)
    
    if len(routes) > 0:
        print("\nAvailable routes:")
        for idx, row in routes.iterrows():
            print(f"  - Route: {row['route_short_name']} ({row['route_id']})")
    else:
        print("\nNo direct routes found between these stops.")
    
    return {
        'start_location': (start_lat, start_lon),
        'end_location': (end_lat, end_lon),
        'nearest_start_stop': nearest_start_stop,
        'nearest_end_stop': nearest_end_stop,
        'start_candidates': start_stops,
        'end_candidates': end_stops,
        'routes': routes
    }

# Example: Plan a journey (Delhi coordinates)
print("\n")
journey = plan_journey(28.7041, 77.1025, 28.6139, 77.2090, df2, df1, df8, df4)
```

```python
# Get detailed path information for a specific route
def get_route_details(trip_id, stop_times_df, stops_df):
    """
    Get all stops and their sequence for a specific trip
    
    Parameters:
    - trip_id: Trip ID to get details for
    - stop_times_df: Stop times dataframe
    - stops_df: Stops dataframe
    
    Returns: DataFrame with stop sequence
    """
    # Get all stops for this trip in order
    trip_stops = stop_times_df[stop_times_df['trip_id'] == trip_id].sort_values('stop_sequence')
    
    # Merge with stop information
    trip_stops_detailed = trip_stops.merge(stops_df[['stop_id', 'stop_name', 'stop_lat', 'stop_lon']], 
                                           on='stop_id', how='left')
    
    return trip_stops_detailed[['stop_sequence', 'stop_id', 'stop_name', 'arrival_time', 'departure_time']]

# Example: Get route details for a trip
if len(journey['routes']) > 0:
    sample_trip = journey['routes'].iloc[0]['trip_id']
    print(f"\nRoute details for Trip {sample_trip}:")
    print("="*60)
    route_details = get_route_details(sample_trip, df1, df2)
    print(route_details.head(10).to_string())
```

```python
import networkx as nx
from datetime import datetime, timedelta

print("\n" + "="*70)
print("BUILDING DIRECTED GRAPHS FOR TRANSIT ROUTING")
print("="*70)

# Function to convert time string to seconds
def time_to_seconds(time_str):
    """Convert HH:MM:SS to seconds"""
    try:
        h, m, s = map(int, str(time_str).split(':'))
        return h * 3600 + m * 60 + s
    except:
        return 0

# Function to build time-weighted directed graph
def build_time_graph(stop_times_df, stops_df, trips_df, routes_df):
    """
    Build a directed graph where edge weights = travel time between stops
    
    Returns: NetworkX DiGraph
    """
    G_time = nx.DiGraph()
    
    # Add all stops as nodes
    for _, stop in stops_df.iterrows():
        G_time.add_node(stop['stop_id'], 
                       name=stop['stop_name'],
                       lat=stop['stop_lat'],
                       lon=stop['stop_lon'])
    
    # Add edges from stop_times data
    # Group by trip to get consecutive stops
    for trip_id in stop_times_df['trip_id'].unique():
        trip_stops = stop_times_df[stop_times_df['trip_id'] == trip_id].sort_values('stop_sequence')
        
        # Get route info for this trip
        trip_route = trips_df[trips_df['trip_id'] == trip_id]
        if len(trip_route) > 0:
            route_id = trip_route['route_id'].values[0]
        else:
            route_id = None
        
        # Connect consecutive stops
        for i in range(len(trip_stops) - 1):
            from_stop = trip_stops.iloc[i]
            to_stop = trip_stops.iloc[i + 1]
            
            # Calculate travel time
            from_time = time_to_seconds(from_stop['departure_time'])
            to_time = time_to_seconds(to_stop['arrival_time'])
            travel_time = max(0, to_time - from_time)  # Handle midnight crossing
            
            # Add edge with travel time as weight
            G_time.add_edge(
                from_stop['stop_id'],
                to_stop['stop_id'],
                weight=travel_time,
                trip_id=trip_id,
                route_id=route_id,
                departure_time=from_stop['departure_time'],
                arrival_time=to_stop['arrival_time']
            )
    
    return G_time

# Build the time-weighted graph
print("\nBuilding time-weighted directed graph...")
G_time = build_time_graph(df1, df2, df8, df4)

print(f"âœ“ Graph created!")
print(f"  â€¢ Nodes (stops): {G_time.number_of_nodes()}")
print(f"  â€¢ Edges (connections): {G_time.number_of_edges()}")

# Show some example edges
print("\nSample edges (stop â†’ stop with travel time):")
sample_edges = list(G_time.edges(data=True))[:5]
for from_stop, to_stop, data in sample_edges:
    from_name = G_time.nodes[from_stop].get('name', f'Stop {from_stop}')
    to_name = G_time.nodes[to_stop].get('name', f'Stop {to_stop}')
    travel_time = data['weight']
    route_id = data.get('route_id', 'N/A')
    print(f"  {from_stop} ({from_name}) â†’ {to_stop} ({to_name}) | Time: {travel_time}s | Route: {route_id}")
```
